#### Feature engineering

```python
df['saledate'] = pd.to_datetime(df['saledate'], dayfirst=True)
```

#### Data preprocessing

```python
y = df['MA'].values
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
result = seasonal_decompose(y, model='additive', period=4)
result.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(y)
print("P-value:", result[1])
```

```python
diff = np.log(y)
```

```python
result2 = adfuller(diff)
print("P-value:", result2[1])
```

```python
from statsmodels.graphics.tsaplots import  plot_acf,plot_pacf
plot_acf(y)
plt.show()

plot_pacf(y)
plt.show()
```

```python
X = df.drop(columns=["MA"])
```

```python
X = pd.get_dummies(df, columns=['type'], drop_first=True).astype(int)
```

```python
X.head()
```

#### Sarimax forecast

```python
from sklearn.model_selection import TimeSeriesSplit
from statsmodels.tsa.statespace.sarimax import SARIMAX

results = []
tscv = TimeSeriesSplit(n_splits=3)

for fold, (train_idx, test_idx) in enumerate(tscv.split(X)):
  X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
  y_train, y_test = y[train_idx], y[test_idx]
  exog_train = pd.DataFrame({'X1' : X_train['bedrooms'], 'X2':X_train['type_unit'] })
  model_sarimax = SARIMAX(
      y_train,
      exog = exog_train,
      order = (1,0,0),
      seasonal_order= (0,1,0,4)).fit()
```

```python
exog_test = pd.DataFrame({'X1' : X_test['bedrooms'], 'X2' : X_test['type_unit']})
```

```python
forecast_sarimax = model_sarimax.forecast(steps=len(y_test), exog=exog_test)
```

```python
forecast_sarimax
```

```python
from sklearn.metrics import mean_absolute_percentage_error

mape = np.mean(np.abs((y_test - forecast_sarimax) / y_test)) * 100
results.append(mape)

print("MAPE per fold:", results)
print("Average MAPE:", np.mean(results))
```

#### Prophet

```python
from prophet import Prophet
```

```python
prophet_df = df.reset_index()
prophet_df.head()
```

```python
prophet_df.dtypes
```

```python
prophet_df.columns = ['ds', 'y', 'X1', 'X2']
```

```python
prophet_df = pd.get_dummies(prophet_df, columns=['X1'], drop_first=True).astype(int)
```

```python
prophet_df['ds'] = pd.to_datetime(prophet_df['ds'], errors='coerce')
```

```python
prophet_df.head()
```

```python
for fold, (train_idx, test_idx) in enumerate(tscv.split(prophet_df)):
    train_df, test_df = prophet_df.iloc[train_idx], prophet_df.iloc[test_idx]
```

```python
train_df.dtypes
```

```python
model_prophet = Prophet(yearly_seasonality=False, weekly_seasonality=False, daily_seasonality=False)
model_prophet.add_seasonality(name ="quarterly", period=365.25/4, fourier_order=5)

results_p = []

for col in train_df.columns:
  if col not in ['ds', 'y']:
    model_prophet.add_regressor(col)

model_prophet.fit(train_df)
```

```python
forecast_prophet = model_prophet.predict(test_df)
```

```python
eval_prophet = pd.DataFrame({
    'ds': test_df['ds'],
    'predicted': forecast_prophet['yhat'].values,
    'Actual' : test_df['y'].values
})
```

```python
eval_prophet['MAPE'] = np.mean(np.abs((eval_prophet['Actual'] - eval_prophet['predicted']) / eval_prophet['Actual'])) * 100
```

```python
mape = eval_prophet['MAPE'].mean()
print(mape)
```

```python
sarimax_resid = y_test - forecast_sarimax
prophet_resid = eval_prophet['Actual']-eval_prophet['predicted']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))

sns.histplot(sarimax_resid, kde=True, color='blue', label='SARIMAX Residuals', stat="density")
sns.histplot(prophet_resid, kde=True, color='red', label='Prophet Residuals', stat="density")

plt.legend()
plt.title("Residuals Histogram with KDE")
plt.show()
```

```python
import statsmodels.api as sm

plt.figure(figsize=(12,5))

# SARIMAX Q-Q
sm.qqplot(sarimax_resid, line='s')
plt.title("SARIMAX Residuals Q-Q Plot")
plt.show()

# Prophet Q-Q
sm.qqplot(prophet_resid, line='s')
plt.title("Prophet Residuals Q-Q Plot")
plt.show()
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
# Ljung-Box test up to lag 10
lb_sarimax = acorr_ljungbox(sarimax_resid, lags=[10], return_df=True)
print(lb_sarimax)
```

```python
lb_prophet = acorr_ljungbox(prophet_resid, lags=[10], return_df=True)
print(lb_prophet)
```
